```python
import openai

def invoke_openai(prompt, streaming=False):

    messages = [{"role": "user", "content": prompt}]

    if streaming:
        response = openai.ChatCompletion.create(
            model="gpt-3.5-turbo",
            messages=messages,
            max_tokens=100,
            stream=True
        )
        print(response)
        for chunk in response:
            print(chunk)
            print(chunk.choices[0].delta.get("content", ""), end="", flush=True)
    else:
        response = openai.ChatCompletion.create(
            model="gpt-3.5-turbo",
            messages=messages,
            max_tokens=100
        )
        return response.choices[0].message["content"]
```

```python

prompt_text = "Tell me a joke about programming."
```

```python
# OpenAI Non-Streaming Response
print("OpenAI Non-Streaming Response:")
print(invoke_openai(prompt_text, streaming=False))
```

```python
print("\nOpenAI Streaming Response:")
invoke_openai(prompt_text, streaming=True)
```

## OPEN AI

```python
from openai import AzureOpenAI
from openai import OpenAI
import os

# Javelin Headers
javelin_api_key ="" # javelin api key
llm_api_key="" # llm api key

javelin_headers = {
    "x-api-key": javelin_api_key,
    "x-javelin-route": "OpenAIInspect"
}

client = OpenAI(api_key=llm_api_key,
                base_url=os.path.join(os.getenv("JAVELIN_BASE_URL"), "v1", "query"),
                default_headers=javelin_headers)
print("client",client)
completion = client.chat.completions.create(
  model="gpt-3.5-turbo",
  messages=[
    {"role": "system", "content": "Hello, you are a helpful scientific assistant."},
    {"role": "user", "content": "What is the chemical composition of sugar?"}
  ]
)

print(completion.model_dump_json(indent=2))

# # Streaming Responses
stream = client.chat.completions.create(
    model="gpt-3.5-turbo",
    messages=[
      {"role": "system", "content": "Hello, you are a helpful scientific assistant."},
      {"role": "user", "content": "What is the chemical composition of sugar?"}
    ],
    stream=True
)
print(stream)
for chunk in stream:
  # print(chunk)
  if chunk.choices:
    print(chunk.choices[0].delta.content or "", end="")
```
