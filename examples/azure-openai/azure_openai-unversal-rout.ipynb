# Azure OpenAI Notebook  
This notebook demonstrates the Azure OpenAI integration with Javelin. 

```python
import os
from openai import AzureOpenAI
from javelin_sdk import JavelinClient, JavelinConfig

print("Azure OpenAI: 1 - Chat completions")

# Retrieve the Javelin API key.
javelin_api_key = "" # add your javelin key here
if not javelin_api_key:
    print("Error: JAVELIN_API_KEY environment variable is not set!")
else:
    print("JAVELIN_API_KEY found.")

# Retrieve the Azure OpenAI API key.
azure_openai_api_key = "" # add your azure openai key here
if not azure_openai_api_key:
    print("Warning: AZURE_OPENAI_API_KEY environment variable is not set!")
else:
    print("AZURE_OPENAI_API_KEY found.")

# Create the AzureOpenAI client.
# IMPORTANT: Specify the azure_deployment parameter to match your Azure model deployment name.
openai_client = AzureOpenAI(
    api_version="2023-07-01-preview",
    azure_endpoint="https://javelinpreview.openai.azure.com",
    api_key=azure_openai_api_key
)

# Initialize the Javelin Client.
config = JavelinConfig(
    base_url="https://api-dev.javelin.live",
    # base_url="http://localhost:8000",
    javelin_api_key=javelin_api_key,
)
client = JavelinClient(config)
client.register_azureopenai(openai_client, route_name="openai-univ")

# Call the Azure OpenAI endpoint for chat completions.
completion = openai_client.chat.completions.create(
    model="gpt-4",  # e.g. gpt-35-instant
    messages=[
        {
            "role": "user",
            "content": "How do I output all files in a directory using Python?",
        },
    ],
)

# Print the response as JSON.
print("Azure OpenAI Response:")
print(completion.to_json())
```
